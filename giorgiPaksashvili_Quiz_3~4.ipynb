```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# datetime ტიპის მონაცემები გადავიყვანე ტექსტურ მონაცემებად და გამოვიტანე მთლიანი ცხრილის პირველი ხუთი სტრიქონის მონაემები
d_parser = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %I-%p')
df = pd.read_csv("./Earthquakes_database.csv")
df.head()
```

```python
# წავშალე დუბლირებული მონაცემები,არსებობის შემთხვევაში,თუმცა როგორც მე დავსკვენი არ არსებობს ამნ ცხრილში იდენტური მონაცემები.
df = df.drop_duplicates()
df
```

```python
# აქ ავარჩიე ის სვეტები რონელიც საჭიროდ არ მომეჩვენა ჩემს ბაზაში და წავშალე.
columns_to_delete = ['Magnitude Error', 'Magnitude Seismic Stations', 'Horizontal Distance', 'Horizontal Error', 'Depth Error']
df = df.drop(columns=columns_to_delete)
print(df.head())
```

```python
# Date სვეტი ავიღე და დავშალე სამ ახალ სვეტად დღეების,თვეეების და წლების მიხედვით.
df[['Day', 'Month', 'Year']] = df['Date'].str.split('/', expand=True)
df.drop('Date', axis=1, inplace=True)
df
```

```python
bins = [0,3,6,9]
labels = ["Weak", "Average", "Major"]
df['Category'] = pd.cut(df['Magnitude'], bins=bins, labels=labels, include_lowest=True)
print(df['Category'])
```

```python
# კოდის ამ წანილში გამოყენებულია ორი ფუნქცია  melt() და pიvot_table() ფუნქცია pivot table ის შესაქმნელად.
meltedData = pd.melt(df, id_vars=['Year', 'Time', 'Type', 'ID', 'Status'],
                    value_vars=['Latitude', 'Longitude', 'Depth', 'Magnitude'],
                    var_name='Variable', value_name='Value')
meltedData
```

```python
# ესეც ჩვენი  pivotTable  ცხრილი
pivotTableData = pd.pivot_table(df, values='Magnitude', index=['Year', 'Time', 'Type', 'ID', 'Status'],
                                 columns=['Magnitude Type'], aggfunc='mean', fill_value=None)

print(pivotTableData)
```

```python
#ეს დიაგრმა გვიჩვენებს წლების მიხედვით მაგნიტუდის  ცვლილებას. 2000 წლიდან უკვე იწყება  მაგნიტუდის მომატება საგრძნობლად.
df.plot(x='Year', y='Magnitude', kind='line', title='Magnitude change over the years')
plt.show()
```

```python
df['Type'].value_counts().plot(kind='bar', title='Earthquake types and their frequency counts')
plt.show()
```

```python
df['Magnitude'].plot(kind='hist', bins=20, title='Magnitude Distribution')
plt.show()
```

```python
df.plot(x='Depth', y='Magnitude', kind='scatter', title='Scatter Plot: Depth vs Magnitude')
plt.show()
```

```python
# აქ ისევ დამჭირდა  'Day', 'Month', and 'Year'  სვეტებუს გაერთიანება ახალი 'Date' სვეტის შესაქმნელად.
df['Date'] = pd.to_datetime(df[['Year', 'Month', 'Day']], format='%Y%m%d', errors='coerce')

# გამოვიტანე კვირის დღეები თოთოეული სტრიქონის მიმართ და დავამატე ახალი სვეტი ამ მონაცემების მიხედვით.
df['Date'].dt.day_name()
df['DayOfWeek'] = df['Date'].dt.day_name()
df.head()
```

```python
# აქ თითოეული სტრიქონი  მონაცემთა ბაზა ID ის მიხედვით დავალაგე. რაგდან ყველას უნიკალური ID მონაცემი აქვს.
df.set_index('ID', inplace=True)
df
```
