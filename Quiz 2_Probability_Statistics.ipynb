## დავალება 4:  კუმულაციური ალბათობა (Cumulative distribution function) - (1 ქულა)
დაითვალეთ შემდეგი კუმულაციური ალბათობა scipy.stats.norm.cdf() ფუნქციის გამოყენებით.
დაითვალეთ კლიენტების რამდენ პროცენტს აქვს შემოსავალი 2000 ევროზე ნაკლები ყოველთვიურად.
norm.cdf(x, mean_val, std_dev_val) ფუნქციის პირველი პარამეტრია სასაზღვრო მნიშვნელობა, მეორე- საშუალო, მესამე -სტდ. გადახრა.

#### მოახდინეთ შედეგების სიტყვიერი ინტერპრეტაცია

```python
from scipy.stats import norm
```

```python

meanIncome = 2500  
stdDevIncome = 500

incomeThreshold = 2000

cumulativeProbability = norm.cdf(incomeThreshold, meanIncome, stdDevIncome)

percentage_below_threshold = cumulativeProbability * 100
print(f"კუმულაციური ალბათობა იმისა, რომ მომხმარებელს აქვს შემოსავალი ნაკლები{incomeThreshold} ევრო არის: {cumulativeProbability:.4f}")
print(f"{incomeThreshold} ევრო არის: {percentage_below_threshold:.2f}%")
```

## დავალება 6: t-test (2 ქულა)
ა) გააკეთეთ t-test ანალიზი (One sample t-test)  რომელიმე სვეტის მიმართ და გამოიყენეთ  ttest_1samp ფუნქცია.

ბ) გააკეთეთ t-test ანალიზი (Two sample t-test) რომელიმე სვეტის მიმართ და გამოიყენეთ  ttest_ind ფუნქცია.

განსაზღვრეთ ნულოვანი და ალტერნატიული ჰიპოთეზა, გამოიყენეთ აღნიშნული ფუნქციები და მოახდინეთ შედეგების სიტყვიერი ინტერპრეტაცია.

```python
from scipy.stats import ttest_1samp,ttest_ind
```

```python
sample_column = 'SampleColumn'

hypothesized_mean = 100


t_statistic, p_value = ttest_1samp(df["Gender"], hypothesized_mean)


print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')


alpha = 0.05
if p_value < alpha:
    print("შერჩევის საშუალო მაჩვენებელი  განსხვავდება ჰიპოთეზირებული პოპულაციის საშუალოსგან.")
else:
    print("არ არის საკმარისი მტკიცებულება განსხვავების დასადგენად.")
```

```python
t_statistic, p_value = ttest_ind(df["Gender"], df["Married"])

# Print the results
print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')

alpha = 0.05
if p_value < alpha:
    print("ორი ჯგუფის საშუალო განსხვავებულია.")
else:
    print("არ არის საკმარისი მტკიცებულება განსხვავებაზე.")
```
