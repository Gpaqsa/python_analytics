```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
myData = pd.read_csv('survey_results_public.csv')
```

```python
myData.head();
myData
```

```python
# ჩემი სასურველი სვეტები და სტრიქონები
myData = myData[['Industry', 'MainBranch', 'Age', 'Employment', 'RemoteWork', 'EdLevel',	'ConvertedCompYearly', 'CodingActivities', 'Frequency_3', 'TimeSearching' ]]
myData
```

```python
# ინდექსირება ცხრილის სვეტების მიმართ.
myData.set_index('Frequency_3')
```

```python
filtData = (myData['Age'] == '25-34 years old') & (myData['Employment'].str.startswith('Employed'))
filtData
```

```python
myData.loc[filtData]
```

```python

myData.sort_values(by=['Employment', 'Industry'], ascending=[True, True], inplace=True)

myData['ConvertedCompYearly'].std()
myData['ConvertedCompYearly'].median()
myData['ConvertedCompYearly'].min()
myData['ConvertedCompYearly'].max()
```

```python
frequency = myData['Frequency_3'].value_counts()

plt.bar(frequency.index, frequency)
plt.xlabel('Frequency Type')
plt.ylabel('Count')
plt.title('Different Frequencys')

plt.show()


   

  
plt.plot(myData['ConvertedCompYearly'], myData['Age'])
plt.title('unemployment rate vs Age')
plt.xlabel('ConvertedCompYearly')
plt.ylabel('Age')
plt.show()
```
